```python
# Add the dependencies
import pandas as pd
import os
import numpy as np
```

```python
# Files to load
school_data_to_load = os.path.join("resources", "schools_complete.csv")
student_data_to_load = os.path.join("resources", "students_complete.csv")
```

```python
### Clean "student_name" in the student_data DF so there are no prefixes and suffixes
# Refer to "cleaning_student_names" notebook for code
```

```python
# 1. Add prefixes and suffixes to a list for removal
prefixes_suffixes = ["Dr. ", "Mr. ", "Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# 2. Iterate through the words in the prefixes_suffixes
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "")
```

```python
# 3. Review "student_name" in the student_data DF
student_data_df
```

```python
### Replace the reading and math scores for 9th graders at Thomas High School with NaN

# 1. Use .loc on the student_data DF to revise the reading scores equal to "NaN" for Thomas HS 9th graders
student_data_df.loc[(student_data_df["grade"] == "9th") & 
                    (student_data_df["school_name"] == "Thomas High School") & 
                   (student_data_df["reading_score"] > 0), "reading_score"] = np.nan
```

```python
# 2. Use .loc on the student_data DF to revise the math scores equal to "NaN" for Thomas HS 9th graders
student_data_df.loc[(student_data_df["grade"] == "9th") & 
                    (student_data_df["school_name"] == "Thomas High School") & 
                   (student_data_df["math_score"] > 0), "math_score"] = np.nan
```

```python
# 3. Display the revised student_data DF
student_data_df
```
