```python
# Adding prefixes and suffixes to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ", "Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# Iterate through the words in the prefixes_suffixes:
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "")
```

```python
# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# Get the total number of students using df.count()
student_count = school_data_complete_df["Student ID"].count()
print(student_count)
```

```python
# Calculate the total number of schools using df.count()
# Using the 'school data' DataFrame to get this count; the complete one
# gets you the same count as student #
school_count = school_data_df["school_name"].count()
print(school_count)
```

```python
# Another method to calculate total number of using is uing the unique() method
# Okay to use the 'complete' DataFrame
school_count_2 = school_data_complete_df["school_name"].unique()
school_count_2
```

```python
# Since the above output is a Python list, we use len() to get the count
len(school_count_2)
```

```python
# Adding len() to above-above formula...
school_count_2 = len(school_data_complete_df["school_name"].unique())
print(school_count_2)
```

```python
# Calculate the total budget using .sum()
total_budget = school_data_df["budget"].sum()
print(total_budget)
```

```python
# Calculate the average reading score using .mean()
average_reading_score = float(school_data_complete_df["reading_score"].mean())
print(average_reading_score)
```

```python
# Caclulate the average math scores using .mean()
average_math_score = float(school_data_complete_df["math_score"].mean())
print(average_math_score)
```

```python
# Determing the passing grade
passing_math = school_data_complete_df["math_score"] >= 70
passing_reading = school_data_complete_df["reading_score"] >= 70
```

```python
# Get all the students who are passing math in a new DataFrame.
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math.head()
```

```python
# Get all the students who are passing reading in a new DataFrame
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading.head()
```

```python
# Get the number of students passing math
passing_math_count = passing_math["student_name"].count()
print(passing_math_count)
```

```python
# Get the number of students passing reading
passing_reading_count = passing_reading["student_name"].count()
print(passing_reading_count)
```

```python
# Get the percentage of students passing math
percent_passing_math = passing_math_count / student_count * 100
print(percent_passing_math)
```

```python
percent_passing_math = passing_math_count / float(student_count) * 100
print(percent_passing_math)
```

```python
# Get the percentage of students passing reading
percent_passing_reading = passing_reading_count / student_count * 100
print(percent_passing_reading)
```

```python
percent_passing_reading = passing_reading_count / float(student_count) * 100
print(percent_passing_reading)
```

```python
# Calculate the number of students who passed both math and reading
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
passing_math_reading.head()
```

```python
# Calculate the number of students who passed both math and reading
overall_passing_math_reading_count = passing_math_reading["student_name"].count()
print(overall_passing_math_reading_count)
```

```python
# Calculate the percentage of students who passed both math and reading
overall_percent_passing_math_reading = overall_passing_math_reading_count / float(student_count) * 100
print(percent_passing_math_reading)
```

```python
# Adding a list of values with keys to create a new DataFrame
district_summary_df = pd.DataFrame( 
    [{"Total Schools": school_count,
     "Total Students": student_count,
     "Total Budget": total_budget,
     "Average Math Score": average_math_score,
     "Average Reading Score": average_reading_score,
     "% Passing Math": percent_passing_math,
     "% Passing Reading": percent_passing_reading,
     "% Overall Passing": overall_percent_passing_math_reading}])
district_summary_df
```

```python
# Format the Total Students to have the comma for a thousands separator
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Students"]
```

```python
# Format the Total Budget to have a comma for a thousands separator, 
# a decimal separator, and a #
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df["Total Budget"]
```

```python
# Format the columns

district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
```

```python
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)

district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)

district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)
```
